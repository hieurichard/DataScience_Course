## 5. Câu hỏi 5: Liệt kê các trận đấu trong bảng  “Match”

```python
'''
dùng lệnh inner join với bảng team 2 lần để trích xuất ra tên của đội chủ nhà và tên của đội khách
bảng Match có quan hệ với bảng Team qua từ khóa home_team_api_id và away_team_api_id
'''
pd.read_sql('''
SELECT m.id, match_api_id,t1.team_long_name home_team_name, t2.team_long_name away_team_name, home_team_goal, away_team_goal
FROM Match m 
INNER JOIN Team as t1 ON m.home_team_api_id = t1.team_api_id
INNER JOIN Team as t2 ON m.away_team_api_id = t2.team_api_id
''',conn)
```

## 7. Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

## 12. Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp theo số lượng giảm dần

```python
# dùng lệnh union để hợp 2 bảng là (mã đội nhà, số bàn thắng chủ nhà) và (mã đội khách, số bàn thắng đội khách )
pd.read_sql('''
SELECT t.team_long_name, sum(total_goals) total_goal
FROM (
    SELECT home_team_api_id team_id, home_team_goal total_goals FROM Match
    UNION ALL
    SELECT away_team_api_id team_id, away_team_goal total_goals FROM Match) m
INNER JOIN Team t ON m.team_id = t.team_api_id
GROUP BY t.team_long_name
ORDER BY total_goal DESC
''',conn)
```

## 13. Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần

```python
# dùng union 2 bảng gồm id đội khách và id đội nhà
pd.read_sql('''
SELECT t.team_long_name, count(*) number_of_match
FROM (
    SELECT home_team_api_id team_id FROM Match
    UNION ALL
    SELECT away_team_api_id team_id FROM Match) m
INNER JOIN Team t ON m.team_id = t.team_api_id
GROUP BY t.team_long_name
ORDER BY number_of_match DESC
''',conn)
```

## 14. Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

```python
'''
sử dụng CASE WHEN trong mục select để phân ra từng điều kiện
dùng union all để hợp bảng 1 (đội nhà, bàn thắng, bàn thua) và bảng 2 (đội khách, bàn thắng, bàn thua)
'''
pd.read_sql('''
SELECT team_long_name,
    (CASE 
    WHEN our_team > rival_team THEN 'win'
    WHEN our_team < rival_team THEN 'lose'
    WHEN our_team = rival_team THEN 'tie'
    END) result , 
    count(*) number_of_time
FROM (
    SELECT home_team_api_id team_id, home_team_goal our_team,away_team_goal rival_team FROM Match
    UNION ALL
    SELECT away_team_api_id team_id, away_team_goal our_team,home_team_goal rival_team FROM Match) m
INNER JOIN Team t ON m.team_id = t.team_api_id
GROUP BY team_long_name, result
''',conn)
```

## 15. Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
'''
dùng case when khi thắng thì sẽ tính là 1, còn lại là 0, tính tổng là ra số trận thắng
'''
pd.read_sql('''
SELECT team_long_name,
    SUM(CASE 
    WHEN our_team > rival_team THEN 1
    ELSE 0
    END) number_of_win
FROM (SELECT home_team_api_id team_id, home_team_goal our_team,away_team_goal rival_team FROM Match
UNION ALL
SELECT away_team_api_id team_id, away_team_goal our_team,home_team_goal rival_team FROM Match) m
INNER JOIN Team t ON m.team_id = t.team_api_id
GROUP BY team_long_name 
ORDER BY number_of_win DESC
LIMIT 10
''',conn)
```

## 16. Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
# tỉ lệ tính trên phần trăm,công thức là số trận từng giải x 100 /  tổng tất cả các trận
# làm tròn đến 2 chữ số sau dấu phẩy, 
pd.read_sql('''
SELECT l.name league_name,  ROUND(count(*)*100.0 /(SELECT COUNT(*) FROM Match),2)  percentage
FROM Match m
INNER JOIN League l
ON m.league_id = l.id
GROUP BY league_name
''',conn)
```

## 17. Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận

```python
# công thức là (tổng bàn thắng chủ nhà + tổng bàn thắng khách)từng giải *100/tổng tất cả bàn thắng
# số phần trăm làm tròn đến 2 chữ số thập phân
pd.read_sql('''
SELECT l.name league_name,  
    ROUND((SUM(home_team_goal)+SUM(away_team_goal))*100.0 /(SELECT SUM(home_team_goal)+SUM(away_team_goal) FROM Match),2)  percentage
FROM Match m
INNER JOIN League l
ON m.league_id = l.id
GROUP BY league_name
''',conn)
```

## 18. Câu hỏi 18: Liệt kê tổng số bàn ghi được của từng giải đấu theo từng mùa giải

```python
# số bàn từng mùa giải từng giải đấu = tổng bàn thắng chủ + bàn thắng khách
pd.read_sql('''
SELECT l.name league_name, season, SUM(home_team_goal)+SUM(away_team_goal) total_goals
FROM Match m
INNER JOIN League l
ON m.league_id = l.id
GROUP BY league_name, season
''',conn)
```

## 21. Câu hỏi 21: Liệt kê cầu thủ có số trận đấu nhiều nhất

## 22. Câu hỏi 22: Liệt kê các cầu thủ có tổng số điểm đánh giá lớn hơn 80

```python
# tính trung bình của các lần overall_rating của các cầu thủ để xem cầu thủ nào lớn hơn 80
pd.read_sql('''
SELECT p.player_name, AVG(overall_rating) average_rating
FROM Player_Attributes pa
INNER JOIN player p ON pa.player_api_id = p.player_api_id
GROUP BY p.player_name
HAVING average_rating > 80
''',conn)
```
